이때 할당할 변수명은 df로 지정하고, 매개변수를 사용하여 'id' 열을 DataFrame의 인덱스로 지정합니다.

```python
df = pd.read_csv('airbnb_과제용.csv', index_col = 'id')
```

(2) 다음 열들이 필요없다고 가정하고, 그 열들을 제거해주려고 합니다.  
적절한 메소드를 사용해서 다음 열들을 제거해주세요. 
- 'room_type', 'description', 'first_review', 'last_review'  


예시)

image.png

```python
df.drop(['room_type', 'description', 'first_review', 'last_review'], axis=1, inplace = True)
```

(3) 칼럼들의 순서를 보기좋게 바꾸고자 합니다. 순서는 다음과 같습니다.  
- 'log_price', 'review_scores_rating', 'number_of_reviews', 'property_type',  
'city', 'latitude', 'longitude', 'amenities'  

그 후 칼럼의 이름을 다음과 같이 바꾸고자 합니다.  
- 'log_price', 'score', 'num_review', 'property_type',  
'city', 'lat', 'lon', 'amenities'  

적절한 메소드를 사용하여 위의 두 작업을 진행해주세요.

예시)

image.png

```python
df = df.reindex(columns=['log_price', 'review_scores_rating', 'number_of_reviews', 'property_type',
'city', 'latitude', 'longitude', 'amenities'])
df
```

```python
df.rename(columns= {'review_scores_rating':'score','number_of_reviews':'num_review','latitude':'lat', 
                   'longitude':'lon'}, inplace = True)
df
```

(4) 결측치가 어느 열에 몇 개나 있는지 확인하고자 합니다.

```python
df.info()
```

(5) 'score' 칼럼에 결측치가 존재하는 것 같습니다.  
결측치가 포함된 행을 모두 삭제해주세요.  

예시)

```python
df.dropna(subset=['score'], inplace = True)
df
```

(6) log_price의 값이 4 이상인 행만 인덱싱하여 df 변수에 할당하세요.   
그리고 num_review 열의 값이 5 이상인 행만 인덱싱하여 df 변수에 할당하세요. 


예시)

image.png

```python
df = df[df['log_price'] >= 4]
df
```

```python
df = df[df['num_review']>=5]
df
```

(7) DataFrame을 인덱스인 'id'를 기준으로 내림차순 정렬해주세요.

예시)

image.png

```python
df = df.sort_index(axis=0, ascending = False)
df
```

(8) lat과 lon 칼럼의 값을 소수점 둘째자리까지 반올림해주세요.  
단, apply 메소드와 lambda 익명함수, numpy의 around 메소드를 모두 활용해주세요.

예시)

image.png

```python
df['lat'] = df['lat'].apply(lambda x : np.around(x,2))
df
```

```python
df['lon'] = df['lon'].apply(lambda x : np.around(x,2))
df
```

(9) 피벗 테이블을 만들고자 합니다.  
pivot이라는 변수를 새로 만들어 이 변수에 피벗테이블을 할당합니다.  
행 방향으로 city, property_type의 종류를 계층적으로 배치합니다.  
요약정보를 뽑아낼 값은 log_price와 score이고,  
요약하는 방법으로는 두 값 모두 max, mean, min입니다.

예시)

image.png

```python
pivot = pd.pivot_table(df, index = ['city', 'property_type'], 
                       values=['log_price','score'], 
                       aggfunc = {"log_price":['max','mean','min'], 'score':['max','mean','min']})
pivot
```

```python
df.to_csv('김종민_기초세션제출.csv', index = True)
```
